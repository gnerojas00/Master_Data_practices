```python
#IMPORTACION DE LIBRERIAS
%matplotlib inline
import random
import os
from openpyxl import Workbook, load_workbook
import pandas as pd
import matplotlib.pyplot as pl
import getpass

#FUNCION MENU PRINCIPAL

def menu_1():

    menu = input("¡Comienza a jugar! ¡Elige tu opcion! \n 1.Partida modo solitario \n 2.Partida 2 jugadores \n 3.Estaditica \n 4.Salir").strip()
    while menu != "":
            
        while not menu.isdigit():
            print('No has ingresado un valor valido, intenta de nuevo')
            menu = input("¡Comienza a jugar! ¡Elige tu opcion! \n 1.Partida modo solitario \n 2.Partida 2 jugadores \n 3.Estaditica \n 4.Salir").strip()
        else:
            menu_n = int(menu)
            if 1 <= menu_n <= 4:
                return menu_n
            else:
                print('El numero ingresado no corresponde a ninguna opcion')
                menu = input("¡Comienza a jugar! ¡Elige tu opcion! \n 1.Partida modo solitario \n 2.Partida 2 jugadores \n 3.Estaditica \n 4.Salir").strip()
    else:
        print('No has ingresado una opcion valida')
        menu = input("¡Comienza a jugar! ¡Elige tu opcion! \n 1.Partida modo solitario \n 2.Partida 2 jugadores \n 3.Estaditica \n 4.Salir").strip()
        

#FUNCION REDIRIGIENDO A OTRAS FUNCIONES DADA LA OPCION DEL MENU ANTERIOR

def menu_dir(menu_n):

    if menu_n == 1:
        opc_1jug()
        archivo_resultados()
        

    elif menu_n == 2:
        opc_2jug()
        archivo_resultados()
        

    elif menu_n == 3:
        mostrar_result()
        mostrar_sort()
        graficos_result()

    else:
        print('Esperamos que te hayas divertido')
        return False


#FUNCION SUBMENU - ELEGIR: DIFICULTAD


def menu_2():
    menu2 = input("Elige el nivel de dificultad: \n 1.Facil(20 intentos) \n 2.Medio(12 intentos) \n 3.Dificil(5 intentos)").strip()
    while menu2 != "":
        while not menu2.isdigit():
            print('No ha ingresado un numero, vuelva a intentarlo')
            menu2 = input("Elige el nivel de dificultad: \n 1.Facil(20 intentos) \n 2.Medio(12 intentos) \n 3.Dificil(5 intentos)").strip()
        
        else: 
            menu2 = int(menu2)
            while 1 <= menu2 <= 3:
               return menu2
            else:
                print('No ha ingresado un numero valido, vuelva a intentarlo')
                menu2 = input("Elige el nivel de dificultad: \n 1.Facil(20 intentos) \n 2.Medio(12 intentos) \n 3.Dificil(5 intentos)").strip()
                        
    else: 
        print('No ha seleccionado ninguna opcion, vuelva a intentarlo')
        menu2 = input("Elige el nivel de dificultad: \n 1.Facil(20 intentos) \n 2.Medio(12 intentos) \n 3.Dificil(5 intentos)").strip()

#FUNCION PARA DETERMINAR EL NUMERO MAX DE VECES PARA LOS BUCLES
def menu_2n(menu2):
    if menu2 == 1:
        return 19 
    elif menu2 == 2:
        return 11 
    else:
        return 4 


#FUNCION VERIFICAR INGRESO DE NUMERO

def pedirverif_n():
    
    n_jugador = getpass.getpass('Adivina un numero entre el 1 y 1000*')

    while not n_jugador.isdigit():
        print('No has ingresado un valor correcto')
        n_jugador = getpass.getpass('Adivina un numero entre el 1 y 1000*')
    else:
        return int(n_jugador)

#FUNCION VERIFICAR NUMERO: 1-1000

def verif_mayormil(numero):
    while not (1 <= int(numero) <= 1000):
        print('El numero que has ingresado no esta en el intervalo (1-1000)')
        numero = getpass.getpass('Adivina un numero entre el 1 y 1000*')
        numero = int(numero)
    else:
        return int(numero)
        

#FUNCION DEL MENU 2 : VERIFICAR QUE ES NUMERO  : ¡NUM QUE DESEAS QUE ADIVINEN!

def pedirverif_n_2():
    
    n_jugador = getpass.getpass('Introduce el numero que deseas que adivinen')

    while not n_jugador.isdigit():
        print('No has ingresado un valor correcto')
        n_jugador = getpass.getpass('Introduce el numero que deseas que adivinen')
    else:
        return int(n_jugador)

#FUNCION 2: VERIFICAR 1-1000 :  ¡NUM QUE DESEAS QUE ADIVINEN!

            
def verif_mayormil_2(numero):
    while not (1 <= numero <= 1000):
        print('El numero que has ingresado no esta en el intervalo (1-1000)')
        numero = getpass.getpass('Introduce el numero que deseas que adivinen')
        while not numero.isdigit():
            print('No has ingresado un valor correcto')
            numero = getpass.getpass('Introduce el numero que deseas que adivinen')
        else:
            while not (1 <= int(numero) <= 1000):
                print('El numero que has ingresado no esta en el intervalo (1-1000)')
                numero = getpass.getpass('Introduce el numero que deseas que adivinen')
            else:
                return int(numero)
    else:
        return numero
        
#FUNCION OPCION MODO: SOLITARIO

def opc_1jug():
    #llamo al menu 2: dificultad 1-3
    select_dif = menu_2() 
    
    #selecciono la dificultad 1-3
    max_intentos = menu_2n(select_dif)
    n_compu = random.randint(1, 1000)
    n_jug = pedirverif_n()
    n_jugv = verif_mayormil(n_jug)
    numero_intentos = 0

    while numero_intentos < max_intentos:
        numero_intentos+= 1
            
        if n_jugv < n_compu:
            print(f'Tu numero de intentos ha sido: {numero_intentos}') 
            print(f'El numero que has insertado es menor que el que estas buscando')
            n_jug = pedirverif_n()
            n_jugv = verif_mayormil(n_jug)
                                                                                
        elif n_jugv > n_compu:
            print(f'Tu numero de intentos ha sido: {numero_intentos}') 
            print(f'El numero que has insertado es mayor que el que estas buscando')
            n_jug = pedirverif_n()
            n_jugv = verif_mayormil(n_jug)

        if n_jugv == n_compu:
            print(f'Has acertado, tu numero de intentos para lograrlo ha sido: {numero_intentos}')
            nombre = input('Escriba su nombre o pseudonimo') 
            Listaresultado = [nombre, '2 jugadores', numero_intentos, 'gano']
            guarda_result(Listaresultado)
            break
    
    else:
        print('Has superado el numero de intentos')
        nombre = input('Escriba su nombre o pseudonimo') 
        Listaresultado = [nombre, 'solitario', numero_intentos, 'perdio']
        guarda_result(Listaresultado)
    
        
#FUNCION OPCION MODO: 2 JUGADORES

def opc_2jug():

    #llamo al menu 2: dificultad 1-3
    select_dif = menu_2() 
    #selecciono la dificultad 1-3
    max_intentos = menu_2n(select_dif)
    jugador1_n = pedirverif_n_2() #CONSTANTE
    jugador1 = verif_mayormil_2(jugador1_n)

    
    jugador2_n = pedirverif_n() #CAMBIARA 
    jugador2 = verif_mayormil(jugador2_n)
    numero_intentos = 0
    

    
    if 1 <= jugador2 <= 1000: 
         
        while (numero_intentos < max_intentos):
            numero_intentos+= 1 
                
            if (jugador1 < jugador2):
                print(f'Tu numero de intentos ha sido: {numero_intentos}') 
                print(f'El numero que has insertado es mayor que el que estas buscando')
                jugador2_n = pedirverif_n() 
                jugador2 = verif_mayormil(jugador2_n)
                                                                                    
            elif (jugador1 > jugador2):
                print(f'Tu numero de intentos ha sido: {numero_intentos}') 
                print(f'El numero que has insertado es menor que el que estas buscando')
                jugador2_n = pedirverif_n() 
                jugador2 = verif_mayormil(jugador2_n)
            
            elif (jugador1 == jugador2):
                print(f'Has acertado, tu numero de intentos para lograrlo ha sido: {numero_intentos}')
                nombre = input('Escriba su nombre o pseudonimo') 
                Listaresultado = [nombre, '2 jugadores', numero_intentos, 'gano']
                guarda_result(Listaresultado)
                break
                
        else:
            print('Has superado el numero de intentos')
            nombre = input('Escriba su nombre o pseudonimo') 
            Listaresultado = [nombre, 'solitario', numero_intentos, 'perdio']
            guarda_result(Listaresultado)
    
    else:
        print('El numero que ha ingresado no esta en el rango 1-1000')
     
#FUNCION: PARA CREAR O ACTIVAR HOJA DE EXCEL

def archivo_resultados():
    
    resultados_jugadores = "results_jug.xlsx"

    if not os.path.exists(resultados_jugadores):
        excel_res = Workbook()
        hoja = excel_res.active
        columnas_nombres = ['Nombre', 'Modo de juego', 'Numero de intentos', 'resultado']
        hoja.append(columnas_nombres)
        excel_res.save(resultados_jugadores)
    
    else:

        excel_res = load_workbook(resultados_jugadores)
        hoja = excel_res.active
        return hoja

#FUNCION PARA GUARDAR LOS DATOS DE LA PARTIDA EN LA HOJA EXCEL

def guarda_result(Listaresultados):
    archivo_resultados()
    excel_res = load_workbook("results_jug.xlsx")
    hoja = excel_res.active
    hoja.append(Listaresultados)
    excel_res.save("results_jug.xlsx")

#FUNCION PARA MOSTRAR LA ULTIMA PARTIDA GUARDADA

def mostrar_result():
    file = pd.read_excel("results_jug.xlsx")
    print('Este ha sido el resultado de su ultima jugada:')
    print(file.tail(1))

#FUNCION PARA VER EL TOTAL DE JUGADORES Y ORGANIZARLOS POR GANAR O PERDER Y NUMERO DE INTENTOS PARA GANAR O PERDER
def mostrar_sort():
    
    file = pd.read_excel("results_jug.xlsx")
    file = file.sort_values(['resultado', 'Numero de intentos'])
    print('Puede compararse con el resto de los jugadores anteriores:')
    print(file)

#FUNCION PARA GRAFICAR RESULTADOS MAS INTERESANTES GLOBALES

def graficos_result():

    file = pd.read_excel(r'results_jug.xlsx')
    
#GRAFICOS PARA: VER TOTALES DE JUGADORES QUE HAN GANADO O PERDIDO
    
#Obtener el total de jugadores que ganaron
    lista_resultados = []
    jug_gano = file[(file['resultado']=='gano')].shape[0]
    
    lista_resultados.append(jug_gano)
    
    
#Obtener el total de jugadores que perdieron
    jug_perdio = file[(file['resultado']=='perdio')].shape[0]
    lista_resultados.append(jug_perdio)
    
    x = ['nº jugadores total con victorias', 'nº jugadores total que han perdido']
    y = lista_resultados
    
    
    pl.bar(x,y, color= ['lightblue', 'salmon'])
    pl.show()
    
    
#GRAFICO PARA VER EL TOTAL DE JUGADORES EN LAS MODALIDADES: SOLITARIO O EN PAREJA
    
    lista_modalidad = []
    n_modosolitario = file[(file['Modo de juego'] == 'solitario')].shape[0]
    
    lista_modalidad.append(n_modosolitario)
    
    n_modo2jugadores = file[(file ['Modo de juego'] == '2 jugadores')].shape[0]
    
    lista_modalidad.append(n_modo2jugadores)
    
    x = ['nº jugadores-modo: solitario', 'nº jugadores-modo: 2 jugadores']
    y = lista_modalidad
    
    pl.bar(x,y, color= ['olive', 'paleturquoise'])
    pl.show()

while True:
    opcion = menu_1()
    if menu_dir(opcion) == False:
        break


menu_n = menu_1()

```
